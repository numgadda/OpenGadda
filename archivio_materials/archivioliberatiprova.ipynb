# Archivio Liberati

After having separated from the full text (*L'ARCHIVIO DEGLI ARCHIVI*) the part regarding this specific archive, I then turned it into a .txt separately and modified with all the specific symbols expressed in the original dissertation.

I then operated onto uploading it and then I split it through the ReGex *'\n'*, creating a list.

```python
import re
with open('/Users/martinapensalfini/Desktop/gadda/archivioliberati2.txt') as f:
  content = f.read()
import pandas as pd

from pandas import read_csv, Series
```

```python
content = content.split("\n")
content = [s for s in content if s]
print(content)
```

From here I created the DataFrame using as cells the different values of the list, the units of the archive and dubbed them under the column name *titolo*

```python
df = pd.DataFrame(content, columns=['titolo'])
print(df)
```

I then extracted the year, month and day

```python
year_pattern = r'\*([0-9]{4}(?:-[0-9]{4})?)\*'

# Extract and create a new column for matched years
df['anno'] = df['titolo'].str.extract(year_pattern)

print(df)
```

```python

pattern_month=r"(?i)\b(?:gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembe|ottobre|novembre|dicembre)\b"
df['mese'] = df['titolo'].str.extract(f'({pattern_month})', expand=False)
print(df)
```

```python
pattern_day = r'\*(\d{1,2})\*'

# Extract the values and concatenate them into a single column
df['giorno'] = df['titolo'].str.extract(pattern_day)

# Print the DataFrame
print(df)
```

Then through a controlled list I extracted the places mentioned in this specific archive.

```python
cities = [
    "Milano", "Longone", "Buenos Aires", "Firenze", "Roma", "Parma", "Capri",
    "Genova", "Bergamo", "Antignano", "S. Margherita Ligure", "Cortina d’Ampezzo",
    "Pocol", "Cinquale", "Chianti", "Isola di Rodi", "Zara", "Venezia",
    "Tripoli", "Sirmione", "Cavalcaselle", "Stresa", "Pegli"
]
pattern = r'\b(?:' + '|'.join(cities) + r')\b'

# Extract the matched cities into a new column 'matched_city'
df['luogo'] = df['titolo'].str.extract(f'({pattern})', expand=False)

# Print the DataFrame
print(df)
```

Finally I saved the DataFrame as an Excel and CSV file

```python
df.to_csv("archivioliberati.csv")
```
