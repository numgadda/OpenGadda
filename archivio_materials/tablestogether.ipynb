I also assured that the year, month and date would be taken as integer and not float.

```python
def extract_numeric(value):
    try:
        return int(float(value))
    except (ValueError, TypeError):
        return None
    
result_df['anno'] = result_df['anno'].apply(extract_numeric)
```

```python
def extract_numeric(value):
    try:
        return int(float(value))
    except (ValueError, TypeError):
        return None
    
result_df['mese'] = result_df['mese'].apply(extract_numeric)
```
