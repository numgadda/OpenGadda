# Archivio Gelli

After having separated from the full text (*L'ARCHIVIO DEGLI ARCHIVI*) the part regarding this specific archive, I then turned it into a .txt separately and modified with all the specific symbols expressed in the original dissertation.

I then operated onto uploading it and then I split it through the ReGex *'F.G.'*, creating a list.

```python
import re
with open('/Users/martinapensalfini/Desktop/gadda/archiviogelli.txt') as f:
  content = f.read()
import pandas as pd

from pandas import read_csv, Series
```

```python
series = content.split('F.G.') # create a list
series = [s for s in series if s] # remove empty strings
for serie in series:
  lines = serie.split("F.G.") # split lines by new line
 


print(lines)
```

From here I created the DataFrame using as cells the different values of the list, the units of the archive and dubbed them under the column name *titolo*

```python
df = pd.DataFrame(series, columns=['titolo'])
print(df)
```

I then created an ID for each unit

```python
df.insert(0, 'archivio', 'Archivio Gelli')

# I created an id for all the units

df_list_id = []
for idx, row in df.iterrows():
    df_list_id.append("unitàgelli-" + str(idx))
df.insert(0, "unitàId", Series(df_list_id, dtype="string"))
```

I extracted the year, month and day, mentioned in the archive.

```python
year_pattern = r'\*([0-9]{4}(?:-[0-9]{4})?)\*'

# Extract and create a new column for matched years
df['anno'] = df['titolo'].str.extract(year_pattern)


print(df)
```

```python

pattern_month=r"(?i)\b(?:gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembe|ottobre|novembre|dicembre)\b"
df['mese'] = df['titolo'].str.extract(f'({pattern_month})', expand=False)
# print(df)
```

```python
pattern_day=r'\(\d{1,2}\)'
df['day'] = df['titolo'].str.extract(f'({pattern_day})', expand=False)
print(df)
```

I then extracted the type of the unit.

```python
new_column_values = []

def categorize_value(value):
    if '(Quaderno)' in value:
        return 'quaderno'
    elif '(Raccoglitore)' in value:
        return 'raccoglitore'
    elif '(Dattiloscritto)' in value:
        return 'dattiloscritto'
    elif '(cartella)' in value.lower():
        return 'cartella'
    elif '(Busta)' in value:
        return 'busta'
    elif '(Involto)' in value:
        return 'involto'
    elif '(estratti o veri e propri fascicoli)' in value:
        return 'estratti e fascicoli'
df['tipo'] = df['titolo'].apply(categorize_value)
print(df)
```

Finally I saved it as an Excel file

```python
df.to_excel("archiviogelli.xlsx")
```
