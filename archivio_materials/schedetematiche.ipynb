```python
import pandas as pd

# Replace 'your_file.xlsx' with your Excel file's path
file_path = '/Users/martinapensalfini/Desktop/gadda/df_date_gadda.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Display the DataFrame
print(df)
```

```python
from docx import Document




# Specified value to search for
specific_value = 'AppLetIt'

# Filter rows where the 'schede tematiche' column contains the specific value
filtered_rows = df[df['schede tematiche'].str.contains(specific_value, na=False)]

# Create a Word document and add content from the filtered DataFrame
doc = Document()
doc.add_heading('Filtered Data', level=1)

# Add the entire rows containing the specific value in 'schede tematiche' to the Word document
for index, row in filtered_rows.iterrows():
    row_data = ', '.join([f"{column}: {value}" for column, value in row.items()])
    doc.add_paragraph(row_data)

# Save the Word document
doc.save('filtered_data.docx')
```
